```python
import  numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report
```

```python
[]# load an existing dataset (Breast Cancer  from sklearn)
from sklearn .datasets import load_breast_cancer
data=load_breast_cancer ()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['malignant']= data.target
```

```python
print(df.info())
print(df.describe())
```

```python

import matplotlib.pyplot as plt
import seaborn as sns

# Box plot (using only two features)
plt.figure(figsize=(8, 5))
sns.boxplot(data=df[['mean radius', 'mean texture']])
plt.title("Boxplot of Key Features")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Heatmap of feature correlations
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.1f')
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

# Create the scaler
scaler = StandardScaler()

# Apply scaling to the feature columns
df[data.feature_names] = scaler.fit_transform(df[data.feature_names])
```

```python
#split dataset
x=df[data.feature_names]
y=df['malignant']
x_train,x_test, y_train,y_test,=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#model Training
model =LogisticRegression()
model.fit(x_train,y_train)
```

# New Section

```python
#model evaluation
predictions = model.predict (x_test)
print("Accuracy :", accuracy_score (y_test ,predictions))
print("Classification Report:\n", classification_report(y_test, predictions))
```

```python
#visualization using matplotlib scatter plot
plt.figure (figsize=(8,6))
plt.scatter(df['mean radius'], df['mean texture'], c=df['malignant'],  cmap='coolwarm',alpha=0.7)
plt.xlabel('Mean Radius')
plt.ylabel('Mean Texture')
plt.title('Breast Cancer Data Visualization')
plt.show()



```
